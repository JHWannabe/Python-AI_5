# **1. Linear Layer**

* 신경망의 가장 기본 구성요소
* Fully-Connected(FC) Layer라고 불리기도 함
* 내부 파라미터에 따른 선형 변환을 수행하는 함수

# **2. Linear Layer 작동 방식**

* 각 입력 노드들에 weight(가중치)를 곱하고 모두 합친 뒤, bias(편향)을 더함
* 행렬곱으로 구현이 가능
* n차원에서 m차원으로 선형 변환 함수
* 이 함수의 파라미터를 조절하면 주어진 입력에 대해 원하는 출력을 만들 수 있음
