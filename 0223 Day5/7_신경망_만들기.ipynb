# **1. Neural Network 구성하기**

```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
```

```python
# Sequential : Neural Network의 각 층을 순서대로 쌓을 수 있는 모델
# Dense : 하나의 뉴런층을 구현
# units : 뉴런 또는 출력 노드의 개수
# input_shape : 입력 데이터의 형태 설정
model = keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
```

# **2. Neural Network 컴파일**

```python
# 모델의 학습에 필요한 손실함수와 optimizer를 설정
# 손실함수 : nn의 예측이 얼마나 잘 맞는지 측정하는 역할
# optimizer : 더 개선도니 예측값을 출력하도록 최적화하는 알고리즘
model.compile(loss='mean_squared_error', optimizer='sgd')
```

# **3. Neural Network 학습**

```python
# fit() 메소드에 의해 주어진 입출력 데이터에 대해 지정한 횟수만큼 학습
# 학습이 진행될 때마다 주어진 입력에 대해 주어진 출력값에 더 가까운 값을 출력
# x, y는 훈련에 사용할 입력과 출력 데이터
# epochs은 주어진 데이터를 한 번 훈련하는 단위
x = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
y = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)
model.fit(x, y, epochs=500)
```

# **4. Neural Network 예측**

```python
# predict() 메소드를 사용하면 특정 입력에 대해 예측값을 얻을 수 있음
pred = model.predict([5.0])
print(pred)
```

# **5. 손실 함수**

```python
model = keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
```

```python
model.compile(loss='mse')
```

```python
pred = model.predict([0])
print(pred)
```

```python
# evaluate() : 손실값을 출력
model.evaluate([0], [[0, 1, 0]])
```
